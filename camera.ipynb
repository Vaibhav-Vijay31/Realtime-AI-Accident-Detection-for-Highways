```python
import cv2
from detection import AccidentDetectionModel
import numpy as np
import os
import random
from twilio.rest import Client

# Twilio setup
account_sid = 'ACc442e60040ccdbaabb89e906ba8b9ad0'  # Your Account SID
auth_token = '531c106c6cccdafd3e04eec902315396'  # Replace with your Auth Token
twilio_number = '+16612506942'  # Replace with your Twilio phone number
to_phone_number = '+919372033921'  # Replace with the destination phone number

client = Client(account_sid, auth_token)

model = AccidentDetectionModel("model.json", 'model_weights.h5')
font = cv2.FONT_HERSHEY_SIMPLEX

# Function to generate a random location (latitude, longitude)
def get_random_location():
    latitude = round(random.uniform(-90, 90), 6)
    longitude = round(random.uniform(-180, 180), 6)
    return latitude, longitude

# Function to make a call using Twilio
def make_call(latitude, longitude):
    call_body = f"Emergency! An accident has been detected at location: Latitude {latitude}, Longitude {longitude}. Please respond immediately."
    call = client.calls.create(
        to=to_phone_number,
        from_=twilio_number,
        twiml=f'<Response><Say>{call_body}</Say></Response>'
    )
    print(f"Started call with SID: {call.sid}")

def startapplication():
    video = cv2.VideoCapture('new.mp4')  # for camera use video = cv2.VideoCapture(0)
    while True:
        ret, frame = video.read()
        if not ret:
            break

        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        roi = cv2.resize(gray_frame, (250, 250))

        pred, prob = model.predict_accident(roi[np.newaxis, :, :])
        if pred == "Accident":
            prob = round(prob[0][0] * 100, 2)
            latitude, longitude = get_random_location()  # Retrieve random location
            
            # Make a call
            make_call(latitude, longitude)

            # Display accident status and probability
            cv2.rectangle(frame, (0, 0), (400, 80), (0, 0, 0), -1)
            cv2.putText(frame, pred + " " + str(prob) + "%", (20, 30), font, 1, (255, 255, 0), 2)
            
            # Display random location
            location_text = f"Lat: {latitude}, Lon: {longitude}"
            cv2.putText(frame, location_text, (20, 60), font, 0.6, (255, 255, 255), 1)

        if cv2.waitKey(33) & 0xFF == ord('q'):
            return
        cv2.imshow('Video', frame)

if __name__ == '__main__':
    startapplication()
```
